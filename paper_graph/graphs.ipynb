```python
import matplotlib.pyplot as plt
import numpy as np

# Labels and data
models = [
    'SVR', 'ESOA', 'ESOA + DE', 'ESOA + Levy Flights', 'GA'
]
mse_values = [
    19.332281781282678,
    8.445239771138247,
    8.441748864913553,
    8.442066355287084,
    8.574932789185658
]
accuracy_scores = [
    0.7363795927585206 * 100,
    0.884838345886572 * 100,
    0.8848859488612755 * 100,
    0.8848816194736474 * 100,
    0.8830698155795527 * 100
]

# Convert MSE to RMSE
rmse_values = [np.sqrt(val) for val in mse_values]

# Adjustable y-axis minimums
ymin_rmse = 2  # Change this as needed
ymin_acc = 70  # Change this as needed

x = np.arange(len(models))  # the label locations

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# RMSE Plot
ax1.bar(models, rmse_values, color='tomato')
ax1.set_ylabel('RMSE')
ax1.set_title('RMSE Across Models')
ax1.set_ylim(bottom=ymin_rmse)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Accuracy Plot
ax2.bar(models, accuracy_scores, color='deepskyblue')
ax2.set_ylabel('Accuracy (%)')
ax2.set_title('Accuracy Across Models')
ax2.set_ylim(bottom=ymin_acc)
ax2.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

SVR vs SVR ESOA

```python
# Indices for SVR, ESOA, GA
selected_indices = [0, 1, 4]
ymin_rmse = min([rmse_values[i] for i in selected_indices]) - 1
ymax_rmse = max([rmse_values[i] for i in selected_indices]) + 0.1
ymin_acc = min([accuracy_scores[i] for i in selected_indices]) - 1
ymax_acc = max([accuracy_scores[i] for i in selected_indices]) + 1

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# RMSE Plot
ax1.bar([models[i] for i in selected_indices], [rmse_values[i] for i in selected_indices], color='tomato')
ax1.set_ylabel('RMSE')
ax1.set_title('RMSE Across Models')
ax1.set_ylim(bottom=ymin_rmse, top=ymax_rmse)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Accuracy Plot
ax2.bar([models[i] for i in selected_indices], [accuracy_scores[i] for i in selected_indices], color='deepskyblue')
ax2.set_ylabel('Accuracy (%)')
ax2.set_title('Accuracy Across Models')
ax2.set_ylim(bottom=ymin_acc, top=ymax_acc)
ax2.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
